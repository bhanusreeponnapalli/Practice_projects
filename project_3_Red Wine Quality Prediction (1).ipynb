```python
y = wine_data['quality'].apply(lambda y_value:1 if y_value>=7 else 0) 
y
```

```python
#finding the variance influence factor in each column 
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)

vif = pd.DataFrame()
vif['vif'] = [variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif['feature'] = x.columns

vif
```

```python
x=x.drop(columns=['fixed acidity','density'],axis=2)
x
```

```python
#finding the best random state
from sklearn.linear_model import LogisticRegression

#metric
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,roc_auc_score

#models
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier

maxAccu =0
maxRs =0
for i in range(1,200):
    x_train,x_test,y_train,y_test =train_test_split(x,y,test_size =.20,random_state=i)
    LR = LogisticRegression()  # 
    LR.fit(x_train,y_train)
    predrs=LR.predict(x_test)
    acc = accuracy_score(y_test,predrs)
    if acc>maxAccu :
        maxAccu =acc
        maxRs =i
print("Best accuracy is:",maxAccu,"On Random state:",maxRs)
```

```python
#lets us train the data
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.75,random_state=50)
```

```python
print('Training Outcome:\n',y_train.value_counts())
```

```python
#their is class imvbalance problem.so,we need to deal with imbalance problem
from imblearn.over_sampling import SMOTE
from collections import Counter
sm =SMOTE()
ove_sample = SMOTE(0.75)
x_train_ns,y_train_ns = ove_sample.fit_resample(x_train,y_train)

print("the number of Classed before fit{}",format(Counter(y_train)))
print("the number of clasees after fit{}",format(Counter(y_train_ns)))

```

```python
#Tunning the parameter using GridSearchCV


#x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=30,random_state=3)
x_train_ns,x_test,y_train_ns,y_test = train_test_split(x,y,train_size=30,random_state=50)

clf = RandomForestClassifier()

params = {'n_estimators':[13,15],
         'criterion':['entropy','gini'],
         'max_depth':[10,15],
         'min_samples_split':[10,11],
         'min_samples_leaf':[5,6]}
grd = GridSearchCV(clf,param_grid=params)
grd.fit(x_train_ns,y_train_ns)

print('best_params= >',grd.best_params_)

#RandomForest Classifier

clf = grd.best_estimator_

clf.fit(x_train_ns,y_train_ns)

y_pred = clf.predict(x_test)

rf_conf_mat = confusion_matrix(y_test,y_pred)

print("\n Confusion_mat:",'\n',rf_conf_mat)
print("\n Accuracy RandomForest Normal===>",accuracy_score(y_test,y_pred))
print('\n')
report =classification_report(y_test,y_pred)
print(report)
```

```python
#Decision tree classifier
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train_ns,y_train_ns)
dt.score(x_test,y_test)

pred_df=dt.predict(x_test)
print("Accuracy",accuracy_score(y_test,pred_df)*100)
print('\n',confusion_matrix(y_test,pred_df))
print(classification_report(y_test,pred_df))
```

```python
#SVC

from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train_ns,y_train_ns)
pred_svc = svc.predict(x_test)
print("Accuracy",accuracy_score(y_test,pred_svc)*100)
print("\n",confusion_matrix(y_test,pred_svc))
print(classification_report(y_test,pred_svc))
```

```python
#cross_validation for RandomForest
from sklearn.model_selection import cross_val_score

crv = cross_val_score(clf,x,y,cv=5)
print("Cross validation score of LogisticRegression Model",crv.mean())
```

```python
#cross_validation for DecisionTree

crv = cross_val_score(dt,x,y,cv=5)
print("Cross validation score of LogisticRegression Model",crv.mean())
```

```python
#cross_validation for SVC

crv = cross_val_score(svc,x,y,cv=5)
print("Cross validation score of LogisticRegression Model",crv.mean())
```

```python
#ROC and AUC plot
from sklearn.metrics import plot_roc_curve

plot_roc_curve(grd.best_estimator_,x_test,y_test)
plt.title("ROC AUC plot")
plt.show()
```
